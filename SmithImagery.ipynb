First we'll read _A Portrait of the Artist as a Young Man_ by James Joyce. If we've already done this there will be a file named "Joyce-Portrait.txt" in the same directory as this notebook, otherwise we'll fetch it from [Gutenberg](http://www.gutenberg.org/files/4217/4217.txt).

```python
import urllib.request, os

filename = "Joyce-Portrait.txt"
if os.path.exists(filename):
    f = open(filename, "r")
    string = f.read()
    f.close()
else:
    url = "http://www.gutenberg.org/files/4217/4217.txt"
    string = urllib.request.urlopen(url).read().decode()
    f = open(filename, "w")
    f.write(string)
    f.close()
```

The Gutenberg file includes a header and a footer. We'll deal with the header easily by virtue of looking for the start of chapter sections, but we need to get rid of the footer.

```python
# trim the end to remove the footer of the Gutenberg file
string = string[0:string.find("End of the Project Gutenberg")]
```

Each chapter is clearly indicated with a label like *Chapter 1*, so we'll go through from 1 to 5 and look for the current chapter header and the next chapter header (or the end of the file if it's the last chapter).

```python
chapterStrings = [] # a list of strings, one for each chapter
for number in range(1,6):
    start = string.find("Chapter "+str(number))
    end = string.find("Chapter "+str(number+1))
    chapterString = string[start:end].strip() # last chapter will have end of -1 but that's ok
    chapterStrings.append(chapterString)
```

Now we want to split up the strings into word tokens so that they can be counted. We'll also do very simple lemmatization of word forms, which will work well for collapsing singular and plural forms ("dog" and "dogs") though less well for verb forms.

Now that we have lemmatized tokens for each chapter, we can proceed to count occurrences of each word from Smith's hand-curated list of imagery words. We don't have an electronic copy of the list, but we do have a scan from this [appendix](http://www.cs.unc.edu/Research/jbsArchive/docs/ImageryAndTheMindOfStephenDedalus/AppendixB/Imagery%20of%20the%20Mind%20Appendix%20B.pdf). The quality of the print is not great so the [OCR](http://en.wikipedia.org/wiki/Optical_character_recognition) is not great, but with a bit of manual proofing, we can produce a fairly good [word list](SmithImageryWordList.txt).

```python
import pandas as pd

smithImageryWords = [line.strip() for line in open('SmithImageryWordList.txt')]
allCounts = []
for smithImageryWord in smithImageryWords:
    counts = [lemmas.count(smithImageryWord) for lemmas in chapterLemmas]
    counts.insert(0, sum(counts))
    allCounts.append(counts)
```

With the per chapter word counts we can now produce a data structure and output that greatly resembles [Smith's own imagery by chapter word list counts](http://www.cs.unc.edu/Research/jbsArchive/docs/ImageryAndTheMindOfStephenDedalus/AppendixB/Imagery%20of%20the%20Mind%20Appendix%20B.pdf).

We have frequency counts by chapter, but what we really want are frequency counts by segments of 500 words. We'll go through each chapter and create buckets of 500 words/lemmas.

```python
import numpy as np

segmentLemmas = [] # all segments from all chapters
for lemmas in chapterLemmas:
    for segs in np.array_split(lemmas, (len(lemmas)/500)):
        segmentLemmas.append(list(segs))
```

Just as we did when counting frequencies by chapter, we now count imagery word frequencies by segments of 500 words.

```python
segmentCounts = []
for lemmas in segmentLemmas:
    count = sum([lemmas.count(smithImageryWord) for smithImageryWord in smithImageryWords])
    segmentCounts.append(count)
```

Let's plot these frequencies.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(segmentCounts)
plt.show()
```

Smith explained in email correspondence that for his graph he didn't use raw counts per segment but instead weighted each count by the logarithm of the frequency of the word in the entitre text – this serves to add more importance to higher frequency terms. Here's Smith's email explanation:

1.  Using the table/matrix of image words, go through the text of Portrait and calculate the  total text frequency for each. For ease of processing, include those numbers in a second column of the image table.
2.  Go through the TABLE again and add a third column that is the logarithm for each of the frequency values.
3.  Go through the TEXT a second time and for each image in a particular 500-word segment, multiply its total text frequency logarithm by the number of times that image appears in the particular segment.
4.  Accumulate the total  of all such values for each 500-word segment and plot those points as your "volume" graph.

The key part is to multiply each count of each imagery word per 500-word segment by the logarithm of that word's frequency in the entire document. Let's do that.

```python
import math

# combine our lemmas per chapter into a single flattened list and only keep imagery words
allSmithImageryLemmas = [item for sublist in chapterLemmas for item in sublist if item in smithImageryWords]

# create a simple count of the imagery words for the entire document
allSmithImageryFreqs = nltk.FreqDist(allSmithImageryLemmas)

# replace the raw count of each word by the logarithmic value of that count
for lemma, count in allSmithImageryFreqs.items():
    allSmithImageryFreqs[lemma] = math.log(count)
```

Let's now go back through our lemmas per 500-word segment and multiply the count by the document's logarithmic count.

```python
weightedSegmentCounts = []
for lemmas in segmentLemmas:
    count = sum([lemmas.count(smithImageryWord) * allSmithImageryFreqs[smithImageryWord] for smithImageryWord in smithImageryWords])
    weightedSegmentCounts.append(count)
```

And now plot the new data.

```python
plt.plot(weightedSegmentCounts)
plt.show()
```
